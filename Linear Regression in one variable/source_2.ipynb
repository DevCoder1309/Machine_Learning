### Linear Regression with one Variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Implementing linear regression model

```python
df = pd.read_csv('./one_var.csv', names=['x', 'y'])
x = df['x']
y = df['y']
df.head()
```

#### Plotting the original data 


```python
plt.scatter(x, y, alpha=0.5)
```

#### Calculating the b0 and b1 for the equation: y_pred = b1 * x + b0

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
num = 0
den = 0
for i in range(len(x)):
    num += (x[i] - x_mean) * (y[i] - y_mean)
    den += (x[i] - x_mean) ** 2
    
b1 = num/den
b0 = y_mean - (b1 * x_mean)
b1, b0
```

```python
y_pred = b1*x + b0
y_pred
```

```python
plt.scatter(x, y, alpha=0.5)
plt.plot(x, y_pred)
```

#### Coefficient of determination => r^2 = 1 - (sum of residues)^2/(sum of total errors)^2

```python
num_r = 0
den_r = 0
for i in range(len(x)):
    num_r += (y[i] - y_pred[i]) ** 2
    den_r += (y[i] - y_mean) ** 2
    
r_2 = 1 - (num_r/den_r)
r_2
```
